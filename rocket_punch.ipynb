```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import pyperclip
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
from pprint import pprint
from tqdm import tqdm, notebook
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import requests
import shutil
```

```python
def save_images(file_path, image_url_list, name_list, num):
    
    img_folder_path = file_path    
    os.makedirs(img_folder_path, exist_ok=True)
    
    for i, image_url in enumerate(image_url_list):
        try:
            response = requests.get(image_url, stream=True)
            with open(img_folder_path + name_list[i].text +'.jpg','wb') as fp:
                shutil.copyfileobj(response.raw, fp)
            print(f'[{num}]번째 page의 [{i+1}]번째 company 로고 저장 완료')            
        except Exception as e:
            print(e)
```

```python
path = r'C:\Users\재욱\Downloads\chromedriver.exe'
driver = webdriver.Chrome(path)
driver.implicitly_wait(2)
companies_url = "https://www.rocketpunch.com/companies/" 
try:
    XPATH = '//*[@id="company-list"]'
    page_num = 1
    
    while True:
        driver.get(f'{companies_url}?page={page_num}')
        division = driver.find_element_by_xpath(XPATH)
        company_image_url_list = []
        unrefined_company_image_url_list = division.find_elements_by_tag_name('img')
        
        for i, unrefined_company_image_url in enumerate(unrefined_company_image_url_list):
            if (unrefined_company_image_url.get_attribute('src') != 'https://static.rocketpunch.com/images/main/onboarding/collaboration.svg') and (unrefined_company_image_url.get_attribute('src') != 'https://static.rocketpunch.com/images/app_download/img-main.jpg'):     
                company_image_url_list.append(unrefined_company_image_url.get_attribute('src'))
        
        company_name_list = division.find_elements_by_tag_name('strong')
        
        save_images("./rocket_punch/", company_image_url_list, company_name_list, page_num)
              
        page_num +=1
         
except Exception as e:
    print(e)
except KeyboardInterrupt:
    print("[KeyboardInterrupt]")
```
